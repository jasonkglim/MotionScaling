# Plot heatmaps for dvrk user performance models

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from performance_models import BayesRegressionPerformanceModel
import os

# Function to add red border to maximum value in each row
def annotate_extrema(data, ax, extrema_type='max'):
    if extrema_type == 'max':
        extrema_index = np.nanargmax(data, axis=1)
        color = 'red'
    if extrema_type == 'min':
        extrema_index = np.nanargmin(data, axis=1)
        color = 'green'
    for i, max_col in enumerate(extrema_index):
        ax.add_patch(plt.Rectangle((max_col, i), 1, 1, fill=False, edgecolor=color, lw=3))

# Loop over users
for user in os.listdir('../dvrk/trial_data'):
	print(user)
	# Load data
	data = pd.read_csv('../dvrk/trial_data/' + user + '/metric_data.csv', index_col=0)
	# print(data)
	# Train model with data
	X = data[['latency', 'scale']]
	metric_list = [header for header in data.columns if header not in ['latency', 'scale']]
	Y = data[metric_list]
	Y_dict = Y.to_dict(orient='list')
	# print(Y_dict)
	model = BayesRegressionPerformanceModel(X, Y_dict, set_poly_transform=2)
	model.train()

	# Predict over same input as training data
	predictions = model.predict(X)
	for metric in metric_list:
		data[metric + '_predicted'] = predictions[metric][1]
	print(data)
	# Plot heatmaps of original and predicted data
	fig, axs = plt.subplots(2, 2, figsize=(18, 6))
	for i, metric in enumerate(metric_list):
		# Original data
		heatmap_data = data.pivot('latency', 'scale', metric)
		sns.heatmap(heatmap_data, ax=axs[0, i], cmap='YlGnBu', annot=True, fmt='.3f')
		axs[0, i].set_title('Original ' + metric)
		annotate_extrema(heatmap_data.values, axs[0, i], extrema_type='min')
		# Predicted data
		heatmap_data = data.pivot('latency', 'scale', metric + '_predicted')
		sns.heatmap(heatmap_data, ax=axs[1, i], cmap='YlGnBu', annot=True, fmt='.3f')
		axs[1, i].set_title('Predicted ' + metric)
		annotate_extrema(heatmap_data.values, axs[1, i], extrema_type='min')
	plt.tight_layout()
	plt.show()

	# Test get_optimal_scale function
	scale_domain = [1, 2, 3, 4]
	for delay in [2, 5]:
		for metric in metric_list:
			print("Optimal Scale for delay " 
			+ str(delay) + ", " 
			+ metric + ": " 
			+ str(model.get_optimal_scale(delay, scale_domain, metric)))
```
